# DK1 vs DK2 day-ahead prices and spread

First exploration: load DK1 + DK2 day-ahead prices and look at the spread
(DK1 − DK2). The two Danish zones sit on different synchronous areas
(DK1 continental, DK2 Nordic), so their prices diverge when the interconnectors
between them bind — see `docs/market-primer.md`.

Run `make fetch START=2026-01-01 END=2026-02-01` first to populate
`data/processed/` (needs an ENTSO-E token), or call the loader directly below.

_Data: ENTSO-E Transparency Platform (CC BY 4.0). Strip heavy output before committing._

```python
import matplotlib.pyplot as plt

from energy_markets.ingest.entsoe import fetch_day_ahead_prices
from energy_markets.plotting.prices import plot_price_series, plot_spread
from energy_markets.transform.spreads import prices_to_wide, zone_spread

START, END = "2026-01-01", "2026-02-01"
ZONES = ["DK1", "DK2"]
```

```python
# Cached after the first call (data/raw/). Needs ENTSOE_API_KEY in .env.
prices = prices_to_wide({z: fetch_day_ahead_prices(z, START, END) for z in ZONES})
prices.describe()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 7), sharex=True)
plot_price_series(prices, ax=ax1, title="DK1 & DK2 day-ahead prices")
plot_spread(zone_spread(prices, "DK1", "DK2"), ax=ax2)
fig.tight_layout()
```
